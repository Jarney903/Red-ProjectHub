```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

data1 = Path('Data/TA_Data_V.2.csv')
TA_df = pd.read_csv(data1)

# Clean

# Outliers - TA Dates 10/1/2018 - 5/1/2019 1859-12038 indexes are turnaround bad data
TA_df = TA_df.drop(index=TA_df.index[1859:12038])
TA_df

# Remove Charge Outliers TA.Charge.BPD > 8000 < 20000
TA_df = TA_df.drop(TA_df[(TA_df['TA.Charge.BPD'] > '5000') & (TA_df['TA.Charge.BPD'] < '30000')].index)
TA_df

# Clean data, drop BAD
TA_df = TA_df[TA_df['TA.Charge.BPD'].str.contains("Bad")==False]
TA_df = TA_df[TA_df['TA.LAB.NC4.VOLPCT'].str.contains("Bad")==False]
TA_df = TA_df[TA_df['TA.OH.PSIG'].str.contains("Bad")==False]
TA_df = TA_df[TA_df['TA.Tray4.F'].str.contains("Bad")==False]
TA_df = TA_df[TA_df['TA.OH.F'].str.contains("Bad")==False]
TA_df = TA_df[TA_df['TA.RECV.PCT'].str.contains("Bad")==False]
TA_df = TA_df[TA_df['TA.RECV.PSIG'].str.contains("Bad")==False]
TA_df = TA_df[TA_df['TA.REFLUX.BPD'].str.contains("Bad")==False]
TA_df = TA_df[TA_df['TA.RECYCLE.TO.CONTACT.BPD'].str.contains("Bad")==False]
TA_df = TA_df[TA_df['TA.RECYCLE.TO.CONTACT.F'].str.contains("Bad")==False]
TA_df = TA_df[TA_df['TA.TRAY39.F'].str.contains("Bad")==False]
TA_df = TA_df[TA_df['TA.TRAY45.F'].str.contains("Bad")==False]
TA_df = TA_df[TA_df['TA.DP.PSIG'].str.contains("Bad")==False]
TA_df = TA_df[TA_df['TA.BTM.PSIG'].str.contains("Bad")==False]
TA_df = TA_df[TA_df['TA.BTM.LEVEL.PCT'].str.contains("Bad")==False]
TA_df = TA_df[TA_df['TA.BTM.REBOIL.OUT.F'].str.contains("Bad")==False]
TA_df = TA_df[TA_df['TA.TOP.REBOIL.OUT.F'].str.contains("Bad")==False]

# Clean data, drop BAD
TA_df = TA_df[TA_df['TA.Charge.BPD'].str.contains("Error")==False]
TA_df = TA_df[TA_df['TA.LAB.NC4.VOLPCT'].str.contains("Error")==False]
TA_df = TA_df[TA_df['TA.OH.PSIG'].str.contains("Error")==False]
TA_df = TA_df[TA_df['TA.Tray4.F'].str.contains("Error")==False]
TA_df = TA_df[TA_df['TA.OH.F'].str.contains("Error")==False]
TA_df = TA_df[TA_df['TA.RECV.PCT'].str.contains("Error")==False]
TA_df = TA_df[TA_df['TA.RECV.PSIG'].str.contains("Error")==False]
TA_df = TA_df[TA_df['TA.REFLUX.BPD'].str.contains("Error")==False]
TA_df = TA_df[TA_df['TA.RECYCLE.TO.CONTACT.BPD'].str.contains("Error")==False]
TA_df = TA_df[TA_df['TA.RECYCLE.TO.CONTACT.F'].str.contains("Error")==False]
TA_df = TA_df[TA_df['TA.TRAY39.F'].str.contains("Error")==False]
TA_df = TA_df[TA_df['TA.TRAY45.F'].str.contains("Error")==False]
TA_df = TA_df[TA_df['TA.DP.PSIG'].str.contains("Error")==False]
TA_df = TA_df[TA_df['TA.BTM.PSIG'].str.contains("Error")==False]
TA_df = TA_df[TA_df['TA.BTM.LEVEL.PCT'].str.contains("Error")==False]
TA_df = TA_df[TA_df['TA.BTM.REBOIL.OUT.F'].str.contains("Error")==False]
TA_df = TA_df[TA_df['TA.TOP.REBOIL.OUT.F'].str.contains("Error")==False]

# Clean data, drop I/O Timeout
TA_df = TA_df[TA_df['TA.Charge.BPD'].str.contains("I/O Timeout")==False]
TA_df = TA_df[TA_df['TA.LAB.NC4.VOLPCT'].str.contains("I/O Timeout")==False]
TA_df = TA_df[TA_df['TA.OH.PSIG'].str.contains("I/O Timeout")==False]
TA_df = TA_df[TA_df['TA.Tray4.F'].str.contains("I/O Timeout")==False]
TA_df = TA_df[TA_df['TA.OH.F'].str.contains("I/O Timeout")==False]
TA_df = TA_df[TA_df['TA.RECV.PCT'].str.contains("I/O Timeout")==False]
TA_df = TA_df[TA_df['TA.RECV.PSIG'].str.contains("I/O Timeout")==False]
TA_df = TA_df[TA_df['TA.REFLUX.BPD'].str.contains("I/O Timeout")==False]
TA_df = TA_df[TA_df['TA.RECYCLE.TO.CONTACT.BPD'].str.contains("I/O Timeout")==False]
TA_df = TA_df[TA_df['TA.RECYCLE.TO.CONTACT.F'].str.contains("I/O Timeout")==False]
TA_df = TA_df[TA_df['TA.TRAY39.F'].str.contains("I/O Timeout")==False]
TA_df = TA_df[TA_df['TA.TRAY45.F'].str.contains("I/O Timeout")==False]
TA_df = TA_df[TA_df['TA.DP.PSIG'].str.contains("I/O Timeout")==False]
TA_df = TA_df[TA_df['TA.BTM.PSIG'].str.contains("I/O Timeout")==False]
TA_df = TA_df[TA_df['TA.BTM.LEVEL.PCT'].str.contains("I/O Timeout")==False]
TA_df = TA_df[TA_df['TA.BTM.REBOIL.OUT.F'].str.contains("I/O Timeout")==False]
TA_df = TA_df[TA_df['TA.TOP.REBOIL.OUT.F'].str.contains("I/O Timeout")==False]

# Create Date_Counter
datecount = []
i=0
for date in TA_df['Date']:
    datecount.append(i)
    i=i+1
TA_df["Date_Count"] = datecount

# Drop Dates
TA_df = TA_df.drop(['Date'], axis=1)
TA_df

# Convert all to FLOAT
TA_df = TA_df.astype("float")

# Binary Qualifiers - Create "OVER Limit = 1" and "Under Limit = 0"
TA_df['Limit'] = [ 1 if x > 13 else 0 for x in TA_df['TA.LAB.NC4.VOLPCT']]

# Drop Lab data
TA_df = TA_df.drop(['TA.LAB.NC4.VOLPCT'], axis = 1)

# Re-Order DF Columns
cols = TA_df.columns.tolist()
cols = cols[-1:] + cols[:-1] # Move last to first
cols = cols[-1:] + cols[:-1] # Move new last to first again
TA_df = TA_df[cols]
TA_df
```

```python
# Build SVM Model
y = TA_df["Limit"]
X = TA_df.drop(["Limit", "Date_Count"], axis=1)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,random_state=0, stratify=y)
## random_state is used to make the data reproducible. 
# Specifying a random state of 1 ensures that the same rows are assigned to train and test sets, respectively.
# A different random state number would distribute different rows of data to the train and test sets.
## Stratifying a dataset divides it proportionally. 
# For example, say that 60% of a dataset belongs to the yellow class, and 40% of it belongs to the purple class. 
# Stratifying it ensures that when the entire dataset is split into training and testing sets, 
# 60% of both will belong to the yellow class, and 40% will belong to the purple class.


## scaling is often necessary with models that are sensitive to large numerical values. 
# This is normally the case with models that measure distances between data points. 
# SVM is one model that usually benefits from scaling.
from sklearn.preprocessing import StandardScaler
# Creating a StandardScaler instance.
scaler = StandardScaler()
# Fitting the Standard Scaler with the training data.
X_scaler = scaler.fit(X_train)
# Scaling the data.
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)


# import the SVC module from Scikit-learn,
from sklearn.svm import SVC
model = SVC(kernel='linear')
model.fit(X_train_scaled, y_train)
y_pred = model.predict(X_test_scaled)
results = pd.DataFrame({"Prediction": y_pred, "Actual": y_test}).reset_index(drop=True)
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred))
results.head(20)
```

```python
# Plot Results
# Create X-Axis for y
plot_df=pd.DataFrame(results)
count = []
i=0
for value in plot_df['Actual']:
    count.append(i)
    i=i+1
plot_df["Count"] = count  

plot_df.head()
plot_df["Actual"] = pd.to_numeric(plot_df["Actual"])
plot_df.dtypes
df = plot_df.head(50)
#Plot
ax = df.plot(kind="scatter", x="Count",y="Actual", color="b", label="Actual Results")
df.plot(kind="scatter" , x="Count",y="Prediction", color="r", label="Predicted Results", alpha = 0.5, ax=ax)


ax.set_xlabel("Days")
ax.set_ylabel("Limit")
ax.set_ylim(-2,2)
plt.show()
```
