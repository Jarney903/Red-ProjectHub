```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

data1 = Path('Data/RX_Data_V.3.csv')
RX_df = pd.read_csv(data1)

# Clean

# Outliers - TA Dates 10/1/2018 - 5/1/2019 1859-12038 indexes are turnaround bad data
RX_df = RX_df.drop(index=RX_df.index[1859:12038])
RX_df

# Remove Charge Outliers TA.Charge.BPD > 8000 < 20000
RX_df = RX_df.drop(RX_df[(RX_df['RX.SOUTH.TA.RECYCLE.CHARGE.BPD'] > '500') & (RX_df['RX.SOUTH.TA.RECYCLE.CHARGE.BPD'] < '30000')].index)
RX_df

# Clean data, drop BAD
RX_df = RX_df[RX_df['TA.OVH.RECYCLE.NC4.VOLPCT'].str.contains("Bad")==False]
RX_df = RX_df[RX_df['RX.TB.RECYCLE.CHARGE.BPD'].str.contains("Bad")==False]
RX_df = RX_df[RX_df['RX.TB.RECYCLE.CHARGE.PCT'].str.contains("Bad")==False]
RX_df = RX_df[RX_df['RX.NORTH.TC.CHARGE.BPD'].str.contains("Bad")==False]
RX_df = RX_df[RX_df['RX.NORTH.TC.CHARGE.CV.PCT'].str.contains("Bad")==False]
RX_df = RX_df[RX_df['RX.NORTH.FLUSH.PSIG'].str.contains("Bad")==False]
RX_df = RX_df[RX_df['RX.NORTH.OUT.F'].str.contains("Bad")==False]
RX_df = RX_df[RX_df['RX.SOUTH.TA.RECYCLE.CHARGE.BPD'].str.contains("Bad")==False]
RX_df = RX_df[RX_df['RX.SOUTH.TA.RECYCLE.CHARGE.PCT'].str.contains("Bad")==False]
RX_df = RX_df[RX_df['RX.SOUTH.TC.CHARGE.BPD'].str.contains("Bad")==False]
RX_df = RX_df[RX_df['RX.SOUTH.TC.CHARGE.CV.PCT'].str.contains("Bad")==False]
RX_df = RX_df[RX_df['RX.SOUTH.FLUSH.PSIG'].str.contains("Bad")==False]
RX_df = RX_df[RX_df['RX.SOUTH.OUT.F'].str.contains("Bad")==False]
RX_df = RX_df[RX_df['RX.SOUTH.WATERRETURN.F'].str.contains("Bad")==False]
RX_df = RX_df[RX_df['RX.SOUTH.WATERRETURN.PH'].str.contains("Bad")==False]
RX_df = RX_df[RX_df['RX.OUT.DELTA.F'].str.contains("Bad")==False]


# Clean data, drop Error
RX_df = RX_df[RX_df['TA.OVH.RECYCLE.NC4.VOLPCT'].str.contains("Error")==False]
RX_df = RX_df[RX_df['RX.TB.RECYCLE.CHARGE.BPD'].str.contains("Error")==False]
RX_df = RX_df[RX_df['RX.TB.RECYCLE.CHARGE.PCT'].str.contains("Error")==False]
RX_df = RX_df[RX_df['RX.NORTH.TC.CHARGE.BPD'].str.contains("Error")==False]
RX_df = RX_df[RX_df['RX.NORTH.TC.CHARGE.CV.PCT'].str.contains("Error")==False]
RX_df = RX_df[RX_df['RX.NORTH.FLUSH.PSIG'].str.contains("Error")==False]
RX_df = RX_df[RX_df['RX.NORTH.OUT.F'].str.contains("Error")==False]
RX_df = RX_df[RX_df['RX.SOUTH.TA.RECYCLE.CHARGE.BPD'].str.contains("Error")==False]
RX_df = RX_df[RX_df['RX.SOUTH.TA.RECYCLE.CHARGE.PCT'].str.contains("Error")==False]
RX_df = RX_df[RX_df['RX.SOUTH.TC.CHARGE.BPD'].str.contains("Error")==False]
RX_df = RX_df[RX_df['RX.SOUTH.TC.CHARGE.CV.PCT'].str.contains("Error")==False]
RX_df = RX_df[RX_df['RX.SOUTH.FLUSH.PSIG'].str.contains("Error")==False]
RX_df = RX_df[RX_df['RX.SOUTH.OUT.F'].str.contains("Error")==False]
RX_df = RX_df[RX_df['RX.SOUTH.WATERRETURN.F'].str.contains("Error")==False]
RX_df = RX_df[RX_df['RX.SOUTH.WATERRETURN.PH'].str.contains("Error")==False]
RX_df = RX_df[RX_df['RX.OUT.DELTA.F'].str.contains("Error")==False]


# Clean data, drop I/O Timeout
RX_df = RX_df[RX_df['TA.OVH.RECYCLE.NC4.VOLPCT'].str.contains("I/O Timeout")==False]
RX_df = RX_df[RX_df['RX.TB.RECYCLE.CHARGE.BPD'].str.contains("I/O Timeout")==False]
RX_df = RX_df[RX_df['RX.TB.RECYCLE.CHARGE.PCT'].str.contains("I/O Timeout")==False]
RX_df = RX_df[RX_df['RX.NORTH.TC.CHARGE.BPD'].str.contains("I/O Timeout")==False]
RX_df = RX_df[RX_df['RX.NORTH.TC.CHARGE.CV.PCT'].str.contains("I/O Timeout")==False]
RX_df = RX_df[RX_df['RX.NORTH.FLUSH.PSIG'].str.contains("I/O Timeout")==False]
RX_df = RX_df[RX_df['RX.NORTH.OUT.F'].str.contains("I/O Timeout")==False]
RX_df = RX_df[RX_df['RX.SOUTH.TA.RECYCLE.CHARGE.BPD'].str.contains("I/O Timeout")==False]
RX_df = RX_df[RX_df['RX.SOUTH.TA.RECYCLE.CHARGE.PCT'].str.contains("I/O Timeout")==False]
RX_df = RX_df[RX_df['RX.SOUTH.TC.CHARGE.BPD'].str.contains("I/O Timeout")==False]
RX_df = RX_df[RX_df['RX.SOUTH.TC.CHARGE.CV.PCT'].str.contains("I/O Timeout")==False]
RX_df = RX_df[RX_df['RX.SOUTH.FLUSH.PSIG'].str.contains("I/O Timeout")==False]
RX_df = RX_df[RX_df['RX.SOUTH.OUT.F'].str.contains("I/O Timeout")==False]
RX_df = RX_df[RX_df['RX.SOUTH.WATERRETURN.F'].str.contains("I/O Timeout")==False]
RX_df = RX_df[RX_df['RX.SOUTH.WATERRETURN.PH'].str.contains("I/O Timeout")==False]
RX_df = RX_df[RX_df['RX.OUT.DELTA.F'].str.contains("I/O Timeout")==False]



# Drop Dates
RX_df = RX_df.drop(['Date'], axis=1)
# Convert all to FLOAT
RX_df = RX_df.astype("float")
RX_df
```

```python
### Prep for Plot Charge ###
RX_df2 = RX_df
# Add Date Count
datecount = []
i=0
for date in RX_df2['RX.SOUTH.TA.RECYCLE.CHARGE.BPD']:
    datecount.append(i)
    i=i+1
RX_df2["Date_Count"] = datecount
RX_df2.plot(x ='Date_Count', y='RX.SOUTH.TA.RECYCLE.CHARGE.BPD', kind = 'line')
plt.show()
```

```python
 ## Separate the Features (X) from the Target (y)
y = RX_df["TA.OVH.RECYCLE.NC4.VOLPCT"]
X = RX_df.drop(["TA.OVH.RECYCLE.NC4.VOLPCT", "Count"], axis = 1)

## Split into train data and test data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,random_state=0)
X_train.shape

# Import Linear Regression Model
from sklearn.linear_model import LinearRegression
model = LinearRegression()

# Train the model
model.fit(X_train, y_train)

# Make Predictions
y_pred = model.predict(X_test)
results = pd.DataFrame({"Prediction": y_pred, "Actual": y_test}).reset_index(drop=True)
results.head(20)

```

```python
# Plot Results
# Create X-Axis for y
plot_df=pd.DataFrame(results)
count = []
i=0
for value in plot_df['Actual']:
    count.append(i)
    i=i+1
plot_df["Count"] = count  
plot_df.head()
plot_df["Actual"] = pd.to_numeric(plot_df["Actual"])
plot_df.dtypes
df = plot_df.head(150)
# Plot
ax = df.plot(kind="scatter", x="Count",y="Actual", color="b", label="Actual Results")
df.plot(x="Count",y="Prediction", color="r", label="Predicted Results", ax=ax)


ax.set_xlabel("Days")
ax.set_ylabel("N-Butane Vol%")
ax.set_ylim(0,25)
plt.show()
```

```python
results_df = pd.DataFrame(results)
variation = abs(results_df["Actual"] - results_df["Prediction"])
results_df['Variation Vol%']=variation
total_variation = results_df['Variation Vol%'].mean()

accuracy = (results_df["Actual"].astype(float)/results_df["Prediction"].astype(float))*100
results_df['Accuracy']=accuracy
total_accuracy = results_df['Accuracy'].astype(float).mean()

# total_accuracy
print(" The average variation is " ,total_variation, "Vol% Points, the average accuracy is " ,total_accuracy, "%")
```
