```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

data1 = Path('Data/TB_Data_V.3.csv')
TB_df = pd.read_csv(data1)

# Clean

# Outliers - TA Dates 10/1/2018 - 5/1/2019 1859-12038 indexes are turnaround bad data
TB_df = TB_df.drop(index=TB_df.index[1859:12038])
TB_df

# Remove Charge Outliers TA.Charge.BPD > 8000 < 20000
TB_df = TB_df.drop(TB_df[(TB_df['TB.CHARGE.BPD'] > '500') & (TB_df['TB.CHARGE.BPD'] < '30000')].index)
TB_df

# Clean data, drop BAD
TB_df = TB_df[TB_df['TA.OVH.RECYCLE.NC4.VOLPCT'].str.contains("Bad")==False]
TB_df = TB_df[TB_df['TB.CHARGE.BPD'].str.contains("Bad")==False]
TB_df = TB_df[TB_df['TB.CHARGE.PREHEAT.COND.LBHR'].str.contains("Bad")==False]
TB_df = TB_df[TB_df['TB.CHARGE.F'].str.contains("Bad")==False]
TB_df = TB_df[TB_df['TB.OVHD.F'].str.contains("Bad")==False]
TB_df = TB_df[TB_df['TB.OVHD.PSIG'].str.contains("Bad")==False]
TB_df = TB_df[TB_df['TB.OVHD.REVEIVER.PSIG'].str.contains("Bad")==False]
TB_df = TB_df[TB_df['TB.OVHD.REVEIVER.F'].str.contains("Bad")==False]
TB_df = TB_df[TB_df['TB.OVHD.REVEIVER.LVL.PCT'].str.contains("Bad")==False]
TB_df = TB_df[TB_df['TB.BTM.PSIG'].str.contains("Bad")==False]
TB_df = TB_df[TB_df['TB.BTM.LVL.PCT'].str.contains("Bad")==False]
TB_df = TB_df[TB_df['TB.DP.PSIG'].str.contains("Bad")==False]
TB_df = TB_df[TB_df['TB.BTM.F'].str.contains("Bad")==False]
TB_df = TB_df[TB_df['TB.REBOIL.OUT.F'].str.contains("Bad")==False]

# Clean data, drop Error
TB_df = TB_df[TB_df['TA.OVH.RECYCLE.NC4.VOLPCT'].str.contains("Error")==False]
TB_df = TB_df[TB_df['TB.CHARGE.BPD'].str.contains("Error")==False]
TB_df = TB_df[TB_df['TB.CHARGE.PREHEAT.COND.LBHR'].str.contains("Error")==False]
TB_df = TB_df[TB_df['TB.CHARGE.F'].str.contains("Error")==False]
TB_df = TB_df[TB_df['TB.OVHD.F'].str.contains("Error")==False]
TB_df = TB_df[TB_df['TB.OVHD.PSIG'].str.contains("Error")==False]
TB_df = TB_df[TB_df['TB.OVHD.REVEIVER.PSIG'].str.contains("Error")==False]
TB_df = TB_df[TB_df['TB.OVHD.REVEIVER.F'].str.contains("Error")==False]
TB_df = TB_df[TB_df['TB.OVHD.REVEIVER.LVL.PCT'].str.contains("Error")==False]
TB_df = TB_df[TB_df['TB.BTM.PSIG'].str.contains("Error")==False]
TB_df = TB_df[TB_df['TB.BTM.LVL.PCT'].str.contains("Error")==False]
TB_df = TB_df[TB_df['TB.DP.PSIG'].str.contains("Error")==False]
TB_df = TB_df[TB_df['TB.BTM.F'].str.contains("Error")==False]
TB_df = TB_df[TB_df['TB.REBOIL.OUT.F'].str.contains("Error")==False]


# Clean data, drop I/O Timeout
TB_df = TB_df[TB_df['TA.OVH.RECYCLE.NC4.VOLPCT'].str.contains("I/O Timeout")==False]
TB_df = TB_df[TB_df['TB.CHARGE.BPD'].str.contains("I/O Timeout")==False]
TB_df = TB_df[TB_df['TB.CHARGE.PREHEAT.COND.LBHR'].str.contains("I/O Timeout")==False]
TB_df = TB_df[TB_df['TB.CHARGE.F'].str.contains("I/O Timeout")==False]
TB_df = TB_df[TB_df['TB.OVHD.F'].str.contains("I/O Timeout")==False]
TB_df = TB_df[TB_df['TB.OVHD.PSIG'].str.contains("I/O Timeout")==False]
TB_df = TB_df[TB_df['TB.OVHD.REVEIVER.PSIG'].str.contains("I/O Timeout")==False]
TB_df = TB_df[TB_df['TB.OVHD.REVEIVER.F'].str.contains("I/O Timeout")==False]
TB_df = TB_df[TB_df['TB.OVHD.REVEIVER.LVL.PCT'].str.contains("I/O Timeout")==False]
TB_df = TB_df[TB_df['TB.BTM.PSIG'].str.contains("I/O Timeout")==False]
TB_df = TB_df[TB_df['TB.BTM.LVL.PCT'].str.contains("I/O Timeout")==False]
TB_df = TB_df[TB_df['TB.DP.PSIG'].str.contains("I/O Timeout")==False]
TB_df = TB_df[TB_df['TB.BTM.F'].str.contains("I/O Timeout")==False]
TB_df = TB_df[TB_df['TB.REBOIL.OUT.F'].str.contains("I/O Timeout")==False]

#Clean data, drop "No Data"
TB_df = TB_df[TB_df['TA.OVH.RECYCLE.NC4.VOLPCT'].str.contains("No Data")==False]
TB_df = TB_df[TB_df['TB.CHARGE.BPD'].str.contains("No Data")==False]
TB_df = TB_df[TB_df['TB.CHARGE.PREHEAT.COND.LBHR'].str.contains("No Data")==False]
TB_df = TB_df[TB_df['TB.CHARGE.F'].str.contains("No Data")==False]
TB_df = TB_df[TB_df['TB.OVHD.F'].str.contains("No Data")==False]
TB_df = TB_df[TB_df['TB.OVHD.PSIG'].str.contains("No Data")==False]
TB_df = TB_df[TB_df['TB.OVHD.REVEIVER.PSIG'].str.contains("No Data")==False]
TB_df = TB_df[TB_df['TB.OVHD.REVEIVER.F'].str.contains("No Data")==False]
TB_df = TB_df[TB_df['TB.OVHD.REVEIVER.LVL.PCT'].str.contains("No Data")==False]
TB_df = TB_df[TB_df['TB.BTM.PSIG'].str.contains("No Data")==False]
TB_df = TB_df[TB_df['TB.BTM.LVL.PCT'].str.contains("No Data")==False]
TB_df = TB_df[TB_df['TB.DP.PSIG'].str.contains("No Data")==False]
TB_df = TB_df[TB_df['TB.BTM.F'].str.contains("No Data")==False]
TB_df = TB_df[TB_df['TB.REBOIL.OUT.F'].str.contains("No Data")==False]



# Drop Dates
TB_df = TB_df.drop(['Date'], axis=1)
# Convert all to FLOAT
TB_df = TB_df.astype("float")
TB_df
```

```python
### Prep for Plot Charge ###
TB_df2 = TB_df
# Add Date Count
datecount = []
i=0
for date in TB_df2['TB.CHARGE.BPD']:
    datecount.append(i)
    i=i+1
TB_df2["Date_Count"] = datecount
TB_df2.plot(x ='Date_Count', y='TB.CHARGE.BPD', kind = 'line')
plt.show()
```

```python
 ## Separate the Features (X) from the Target (y)
y = TB_df["TA.OVH.RECYCLE.NC4.VOLPCT"]
X = TB_df.drop(["TA.OVH.RECYCLE.NC4.VOLPCT", "Count"], axis = 1)

## Split into train data and test data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,random_state=0)
X_train.shape

# Import Linear Regression Model
from sklearn.linear_model import LinearRegression
model = LinearRegression()

# Train the model
model.fit(X_train, y_train)

# Make Predictions
y_pred = model.predict(X_test)
results = pd.DataFrame({"Prediction": y_pred, "Actual": y_test}).reset_index(drop=True)
results.head(20)

```

```python
# Plot Results
# Create X-Axis for y
plot_df=pd.DataFrame(results)
count = []
i=0
for value in plot_df['Actual']:
    count.append(i)
    i=i+1
plot_df["Count"] = count  
plot_df.head()
plot_df["Actual"] = pd.to_numeric(plot_df["Actual"])
plot_df.dtypes
df = plot_df.head(150)
# Plot
ax = df.plot(kind="scatter", x="Count",y="Actual", color="b", label="Actual Results")
df.plot(x="Count",y="Prediction", color="r", label="Predicted Results", ax=ax)


ax.set_xlabel("Days")
ax.set_ylabel("N-Butane Vol%")
ax.set_ylim(0,25)
plt.show()
```

```python
results_df = pd.DataFrame(results)
variation = abs(results_df["Actual"] - results_df["Prediction"])
results_df['Variation Vol%']=variation
total_variation = results_df['Variation Vol%'].mean()

accuracy = (results_df["Actual"].astype(float)/results_df["Prediction"].astype(float))*100
results_df['Accuracy']=accuracy
total_accuracy = results_df['Accuracy'].astype(float).mean()

# total_accuracy
print(" The average variation is " ,total_variation, "Vol% Points, the average accuracy is " ,total_accuracy, "%")
```
