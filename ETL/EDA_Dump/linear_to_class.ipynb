```python
# imports
import pandas as pd
import numpy as np
```

```python
# load in file and convert to dataframe

file = pd.read_csv("Resources/ta_pass_03-Copy1.csv", index_col=[0])

linear_df = pd.DataFrame(file)
linear_df = linear_df.drop(columns='Time')
```

```python
linear_df.head()
```

```python
linear_df['TA.OVH.RECYCLE.NC4.VOLPCT'].describe()
```

```python
# find values above 40 and nullify
linear_df.loc[linear_df['TA.OVH.RECYCLE.NC4.VOLPCT'] >= 40] = np.nan
```

```python
# check nulls
linear_df.isnull().sum()
```

```python
# drop nulls
linear_df = linear_df.dropna()
```

NC4 values between 0-39

- 15% + is unsafe (>=15)
- 14% - is okay (<15)

- 0 = OK
- 1 = Unsafe

```python
linear_df['Class'] = [1 if x >= 15 else 0 for x in linear_df['TA.OVH.RECYCLE.NC4.VOLPCT']]
```

- normal butane level 5-7%

- normal (<11) = 0

- 12-14 approaching unsafe = 1

- 15+ = 2

```python
linear_df['Class'] = [2 if x >= 15 else 1 if x < 15 else 0 for x in linear_df['TA.OVH.RECYCLE.NC4.VOLPCT']]
```

```python
linear_df['Class'].unique()
```

```python
normal = linear_df.loc[linear_df['Class'] == 0].count()[17]
approach = linear_df.loc[linear_df['Class'] == 1].count()[17]
unsafe = linear_df.loc[linear_df['Class'] == 2].count()[17]

print(f'Normal: {normal}')
print(f'Approaching Unsafe: {approach}')
print(f'Unsafe: {unsafe}')
```
