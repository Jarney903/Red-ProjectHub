```python
# trying to find actual text strings and in which columns they are in
import re
names_columns = ['TA.LAB.NC4.VOLPCT',
         'TA.Charge.BPD',
      'TA.OH.PSIG',
      'TA.Tray4.F',
      'TA.OH.F',
      'TA.RECV.PCT',
      'TA.RECV.PSIG',
      'TA.REFLUX.BPD',
      'TA.RECYCLE.TO.CONTACT.BPD',
      'TA.RECYCLE.TO.CONTACT.F',
      'TA.TRAY39.F',
      'TA.TRAY45.F',
      'TA.DP.PSIG',
      'TA.BTM.PSIG',
      'TA.BTM.LEVEL.PCT',
      'TA.BTM.REBOIL.OUT.F',
      'TA.TOP.REBOIL.OUT.F']

for name in names_columns:
    ids = play_df[name].str.contains('Bad$|Bad-|Bad|Error$|Error-|Error|I/O Timeout%|I/O Timeout-|I/O Timeout ', flags = re.IGNORECASE, regex = True, na = False)
    print(ids.describe())
```

```python
for name in names_columns:
    #print(name)
    play_clean = play_df[play_df[name].str.contains('Bad')==False]
    play_clean = play_df[play_df[name].str.contains('Error')==False]
    play_clean = play_df[play_df[name].str.contains('I/O Timeout')==False]
    
len(play_clean)
```
