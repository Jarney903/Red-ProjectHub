```python
import pandas as pd
import numpy as np
```

```python
file = pd.read_csv("Resources/TA_Data_V.2.csv")
ta_df = pd.DataFrame(file)
ta_df
```

```python
pd.to_datetime(ta_df['Date'])
```

```python
dates = ta_df['Date']
dates.to_csv(dates.csv)
```

```python
names_columns = ['TA.LAB.NC4.VOLPCT',
         'TA.Charge.BPD',
      'TA.OH.PSIG',
      'TA.Tray4.F',
      'TA.OH.F',
      'TA.RECV.PCT',
      'TA.RECV.PSIG',
      'TA.REFLUX.BPD',
      'TA.RECYCLE.TO.CONTACT.BPD',
      'TA.RECYCLE.TO.CONTACT.F',
      'TA.TRAY39.F',
      'TA.TRAY45.F',
      'TA.DP.PSIG',
      'TA.BTM.PSIG',
      'TA.BTM.LEVEL.PCT',
      'TA.BTM.REBOIL.OUT.F',
      'TA.TOP.REBOIL.OUT.F']
```

```python
for name in names_columns:
    ta_df.loc[ta_df[name] == "I/O Timeout"] = np.nan
    ta_df.loc[ta_df[name] == "Error"] = np.nan
    ta_df.loc[ta_df[name] == "Bad"] = np.nan
    ta_df.loc[ta_df[name] == "Bad Input"] = np.nan

# check to see if these have been replaced properly, should equal 1056 
ta_df.isnull().sum()
```

```python
ta_df = ta_df.drop(columns=['Date'])
ta_df.head(1)
```

```python
ta_df = ta_df.astype(dict.fromkeys(ta_df.columns[0:17], float))
```
