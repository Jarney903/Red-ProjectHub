Linear Regression Test

```python
 ## Separate the Features (X) from the Target (y)
y = TA_df["TA.LAB.NC4.VOLPCT"]
X = TA_df.drop(["TA.LAB.NC4.VOLPCT", "Date"], axis = 1)
#print(y)
#print(X)
#X.describe()
```

```python
## Split into train data and test data
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,random_state=0)
X_train.shape
```

```python
# Import Linear Regression Model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# Train the model
model.fit(X_train, y_train)
```

```python
# Make Predictions
y_pred = model.predict(X_test)
results = pd.DataFrame({"Prediction": y_pred, "Actual": y_test}).reset_index(drop=True)
results.head(20)
```

```python
# Create X-Axis for y

plot_df=pd.DataFrame(results)
count = []
i=0
for value in plot_df['Actual']:
    count.append(i)
    i=i+1
plot_df["Count"] = count  

plot_df.head()
plot_df["Actual"] = pd.to_numeric(plot_df["Actual"])
plot_df.dtypes
df = plot_df.head(150)
df
```

```python
ax = df.plot(kind="scatter", x="Count",y="Actual", color="b", label="Actual Results")
df.plot(x="Count",y="Prediction", color="r", label="Predicted Results", ax=ax)


ax.set_xlabel("Days")
ax.set_ylabel("N-Butane Vol%")
ax.set_ylim(0,25)
plt.show()
```

```python
results_df = pd.DataFrame(results)
variation = (results_df['Actual']-(abs(results_df['Prediction'])- results_df['Actual']))/results_df['Actual']*100
results_df['Accuracy%']=variation
results_df
total_variation = results_df['Accuracy%'].mean()
print(" The total accuracy is " ,total_variation, "%")
```
