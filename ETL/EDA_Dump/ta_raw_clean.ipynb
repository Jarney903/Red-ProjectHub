```python
file2 = pd.read_csv("checkpoint_filter.csv", index_col=[0])
checkpoints = pd.DataFrame(file2)
checkpoints
```

```python
ta_new = checkpoints.loc[checkpoints['Time'] != 0]
```

```python
ta_new
```

```python
ta_new = ta_new.reset_index()
```

```python
# cleaning out text strings
names_columns = ['TA.Charge.BPD',
      'TA.OH.PSIG',
      'TA.Tray4.F',
      'TA.OH.F',
      'TA.RECV.PCT',
      'TA.RECV.PSIG',
      'TA.REFLUX.BPD',
      'TA.RECYCLE.TO.CONTACT.BPD',
      'TA.RECYCLE.TO.CONTACT.F',
      'TA.TRAY39.F',
      'TA.TRAY45.F',
      'TA.DP.PSIG',
      'TA.BTM.PSIG',
      'TA.BTM.LEVEL.PCT',
      'TA.BTM.REBOIL.OUT.F',
      'TA.TOP.REBOIL.OUT.F']

for name in names_columns:
    ta_new.loc[ta_new[name] == "I/O Timeout"] = np.nan
    ta_new.loc[ta_new[name] == "Error"] = np.nan
    ta_new.loc[ta_new[name] == "Bad"] = np.nan
    ta_new.loc[ta_new[name] == "Bad Input"] = np.nan
```

```python
ta_new.isnull().sum()
```

```python
ta_new = ta_new.dropna()
ta_new
```

```python
ta_clean = ta_new.drop(columns=['index', 'Time'])
```

```python
ta_clean = ta_clean.astype(dict.fromkeys(ta_clean.columns[1:17], float))
ta_clean.dtypes
```
