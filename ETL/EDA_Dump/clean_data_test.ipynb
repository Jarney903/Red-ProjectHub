# Clean Lab Data

```python
# convert Date column from object to datetime
lab_data['Date'] = pd.to_datetime(lab_data['Date'])
lab_data.head(1)
```

```python
# add hour column to sort out invalid checkpoint hours
lab_data['Hour'] = pd.to_datetime(lab_data['Date']).dt.hour
lab_data.head(1)
```

```python
# keep only rows that have valid hour checkpoints
lab_data = lab_data.loc[(lab_data['Hour'] == 6) | (lab_data['Hour'] == 13)]
lab_data
```

```python
# store start and end dates of turnaround as variables
start = "2018-10-12 13:00:00"
end = "2019-03-29 6:00:00"

# turn datetimes between start and end into nulls
lab_data.loc[(lab_data['Date'] > start) & (lab_data['Date'] <= end)] = np.nan
```

```python
# for loop to nullify text strings
for name in lab_columns:
    lab_data.loc[lab_data[name] == "I/O Timeout"] = np.nan
    lab_data.loc[lab_data[name] == "Error"] = np.nan
    lab_data.loc[lab_data[name] == "Bad"] = np.nan
    lab_data.loc[lab_data[name] == "Bad Input"] = np.nan
```

```python
# check nulls
lab_data.isnull().sum()
```

```python
# drop nulls from dataframe
lab_data = lab_data.dropna()
lab_data
```

```python
# check data types
lab_data.dtypes
```

```python
# convert object types to float types
lab_data['TA.OVH.RECYCLE.NC4.VOLPCT'] = lab_data['TA.OVH.RECYCLE.NC4.VOLPCT'].astype(float)
lab_data.dtypes
```

```python
# reset index 
lab_data = lab_data.reset_index()
#lab_data
```

```python
# define find outlier function
def find_outliers(df):
    q1 = df.quantile(0.25)
    q3 = df.quantile(.75)
    iqr = q3-q1
    lower_lim = q1 - (1.5 * iqr)
    upper_lim = q3 + (1.5 * iqr)
    outliers = df[((df<lower_lim) | (df>upper_lim))]
    return outliers
```

```python
# run function on lab data
lab_outliers = find_outliers(lab_data['TA.OVH.RECYCLE.NC4.VOLPCT'])
lab_outliers.describe()
```

Through discussion, we have determined that values above 40 will be considered invalid and should be removed. All other values are within reasonable measure, regardless of statistical outliers.

```python
# find invalid values and nullify
lab_data.loc[lab_data['TA.OVH.RECYCLE.NC4.VOLPCT'] >= 40] = np.nan
```

```python
# drop the nulls
lab_data = lab_data.dropna()
```

```python
# rerun outlier function
lab_outliers = find_outliers(lab_data['TA.OVH.RECYCLE.NC4.VOLPCT'])
lab_outliers.describe()
```
